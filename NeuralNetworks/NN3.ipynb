# square-large (MSE: 1)

```python
square_large_train = pd.read_csv(
    "https://raw.githubusercontent.com/nizwant/miowid/main/data/regression/square-large-training.csv",
    index_col=0,
)
square_large_test = pd.read_csv(
    "https://raw.githubusercontent.com/nizwant/miowid/main/data/regression/square-large-test.csv",
    index_col=0,
)
```

```python
mean = square_large_train.mean()
std = square_large_train.std()
square_large_train = (square_large_train - mean) / std
square_large_test = (square_large_test - mean) / std
```

```python
plt.scatter(square_large_train.x, square_large_train.y, label="train")
plt.scatter(square_large_test.x, square_large_test.y, label="test")
```

```python
neural = nn.NeuralNetwork([10], 1, 1)
mse, mse_after_epoch = neural.adam(
    square_large_train["x"],
    square_large_train["y"],
    learning_rate=0.001,
    batch_size=25,
    max_num_epoch=5000,
    silent=True,
)
neural.mean_squared_error(square_large_test["x"], square_large_test["y"]) * std[
    "y"
] ** 2
```

```python
plt.plot(mse)
plt.title("Mean Squared Error")
plt.xlabel("Epoch")
plt.ylabel("MSE")
```

## Test

```python
y = []
for i in square_large_test["x"]:
    y.append(neural.forward(i))
plt.scatter(square_large_test["x"], y, c="red")
plt.scatter(square_large_test["x"], square_large_test["y"], c="blue")
plt.legend(["Prediction", "True"])
neural.mean_squared_error(square_large_test["x"], square_large_test["y"]) * std[
    "y"
] ** 2
```

## Train

```python
y = []
for i in square_large_train["x"]:
    y.append(neural.forward(i))
plt.scatter(square_large_train["x"], y, c="red")
plt.scatter(square_large_train["x"], square_large_train["y"], c="blue")
plt.legend(["Prediction", "True"])
neural.mean_squared_error(square_large_train["x"], square_large_train["y"]) * std[
    "y"
] ** 2
```

# Comparison of optimizer algorithms  

```python
square_large_train = pd.read_csv(
    "https://raw.githubusercontent.com/nizwant/miowid/main/data/regression/square-large-training.csv",
    index_col=0,
)
square_large_test = pd.read_csv(
    "https://raw.githubusercontent.com/nizwant/miowid/main/data/regression/square-large-test.csv",
    index_col=0,
)
mean = square_large_train.mean()
std = square_large_train.std()
square_large_train = (square_large_train - mean) / std
square_large_test = (square_large_test - mean) / std
```

```python
neural_adam = nn.NeuralNetwork([10], 1, 1)
_, mse_after_epoch_adam = neural_adam.adam(
    square_large_train["x"],
    square_large_train["y"],
    learning_rate=0.01,
    batch_size=25,
    max_num_epoch=500,
    silent=True,
)
neural_adam.mean_squared_error(square_large_test["x"], square_large_test["y"]) * std[
    "y"
] ** 2
```

```python
neural_moment = nn.NeuralNetwork([10], 1, 1)
_, mse_after_epoch_moment = neural_moment.mini_batch_gradient_descent_with_momentum(
    square_large_train["x"],
    square_large_train["y"],
    learning_rate=0.01,
    batch_size=25,
    max_num_epoch=500,
    silent=True,
)
neural_moment.mean_squared_error(square_large_test["x"], square_large_test["y"]) * std[
    "y"
] ** 2
```

```python
neural_rms_prop = nn.NeuralNetwork([10], 1, 1)
_, mse_after_epoch_rms_prop = neural_rms_prop.rmsprop(
    square_large_train["x"],
    square_large_train["y"],
    learning_rate=0.01,
    batch_size=25,
    max_num_epoch=500,
    silent=True,
)
neural_rms_prop.mean_squared_error(
    square_large_test["x"], square_large_test["y"]
) * std["y"] ** 2
```

```python
neural_stochastic = nn.NeuralNetwork([10], 1, 1)
_, mse_after_epoch_stochastic = (
    neural_stochastic.mini_batch_gradient_descent(
        square_large_train["x"],
        square_large_train["y"],
        learning_rate=0.01,
        batch_size=2,
        max_num_epoch=500,
        silent=True,
    )
)
neural_stochastic.mean_squared_error(
    square_large_test["x"], square_large_test["y"]
) * std["y"] ** 2
```

```python
neural_mini_batch = nn.NeuralNetwork([10], 1, 1)
_, mse_after_epoch_mini_batch = neural_mini_batch.mini_batch_gradient_descent(
    square_large_train["x"],
    square_large_train["y"],
    learning_rate=0.01,
    batch_size=25,
    max_num_epoch=500,
    silent=True,
)
neural_mini_batch.mean_squared_error(
    square_large_test["x"], square_large_test["y"]
) * std["y"] ** 2
```

```python
neural_full_batch = nn.NeuralNetwork([10], 1, 1)
_, mse_after_epoch_full_batch = neural_full_batch.mini_batch_gradient_descent(
    square_large_train["x"],
    square_large_train["y"],
    learning_rate=0.1,
    batch_fraction=1.0,
    max_num_epoch=500,
    silent=True,
)
neural_full_batch.mean_squared_error(
    square_large_test["x"], square_large_test["y"]
) * std["y"] ** 2
```

```python
mse_after_epoch_adam = np.array(mse_after_epoch_adam) * std["y"] ** 2
mse_after_epoch_moment = np.array(mse_after_epoch_moment) * std["y"] ** 2
mse_after_epoch_rms_prop = np.array(mse_after_epoch_rms_prop) * std["y"] ** 2
mse_after_epoch_stochastic = np.array(mse_after_epoch_stochastic) * std["y"] ** 2
mse_after_epoch_mini_batch = np.array(mse_after_epoch_mini_batch) * std["y"] ** 2
mse_after_epoch_full_batch = np.array(mse_after_epoch_full_batch) * std["y"] ** 2
```

```python
# Plotting
plt.plot(mse_after_epoch_adam, label="Adam")
plt.plot(mse_after_epoch_moment, label="Momentum")
plt.plot(mse_after_epoch_rms_prop, label="RMSProp")
plt.plot(mse_after_epoch_stochastic, label="Stochastic GD")
plt.plot(mse_after_epoch_mini_batch, label="Mini Batch GD")
plt.plot(mse_after_epoch_full_batch, label="Full Batch GD")
plt.title("Mean Squared Error")
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.legend()
```

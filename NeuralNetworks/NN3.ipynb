# square-large (MSE: 1)

```python
square_large_train = pd.read_csv(
    "https://raw.githubusercontent.com/nizwant/miowid/main/data/regression/square-large-training.csv",
    index_col=0,
)
square_large_test = pd.read_csv(
    "https://raw.githubusercontent.com/nizwant/miowid/main/data/regression/square-large-test.csv",
    index_col=0,
)
```

```python
mean = square_large_train.mean()
std = square_large_train.std()
square_large_train = (square_large_train - mean) / std
square_large_test = (square_large_test - mean) / std
```

```python
plt.scatter(square_large_train.x, square_large_train.y, label="train")
plt.scatter(square_large_test.x, square_large_test.y, label="test")
```

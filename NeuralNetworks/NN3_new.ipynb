# multimodal-large (MSE: 9)

```python
multimodal_large_train = pd.read_csv(
    "https://raw.githubusercontent.com/nizwant/miowid/main/data/regression/multimodal-large-training.csv"
)
multimodal_large_test = pd.read_csv(
    "https://raw.githubusercontent.com/nizwant/miowid/main/data/regression/multimodal-large-test.csv",
)
```

```python
mean = multimodal_large_train.mean()
std = multimodal_large_train.std()
multimodal_large_train = (multimodal_large_train - mean) / std
multimodal_large_test = (multimodal_large_test - mean) / std
```

```python
plt.scatter(multimodal_large_train.x, multimodal_large_train.y, label="train")
plt.scatter(multimodal_large_test.x, multimodal_large_test.y, label="test")
```

```python
nural = nn.NeuralNetwork(optimizer="adam")
nural.add_layer(
    layer.Layer(
        1,
        10,
        activation="sigmoid",
        weight_initialization="he",
    )
)
nural.add_layer(
    layer.Layer(
        10,
        5,
        activation="sigmoid",
        weight_initialization="he",
    )
)
nural.add_layer(
    layer.Layer(
        5,
        1,
        activation="linear",
    )
)

df = pd.read_csv(
    "https://raw.githubusercontent.com/nizwant/miowid/main/data/regression/steps-large-training.csv"
)

mean = df.mean()
std = df.std()
df = (df - mean) / std
mse_l = nural.train(df["x"], df["y"], batch_size=64, max_num_epoch=2500, learning_rate=0.01, silent=False)
mse = nural.calculate_cost(df["x"], df["y"])
print(f"Mean Squared Error: {mse * std['y']**2}")
y = []
for i in df["x"]:
    y.append(nural._forward(np.array(i)))
plt.scatter(df["x"], y, c="red")
plt.scatter(df["x"], df["y"], c="blue")
plt.legend(["Prediction", "True"])
plt.show()
```
